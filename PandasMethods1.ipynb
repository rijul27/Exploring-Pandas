```python
#  set one or more columns as the DataFrame's index. In this specific case,
# it's taking the island column from your DataFrame df and making it the new
# index of the DataFrame. and for permanently modification use inplace

# To prevent KeyError if 'island' is already the index (and thus not a column),
# we first reset the index to make 'island' a column again, if it's not already a column.
if 'island' not in df.columns:
    df.reset_index(inplace=True)
df.set_index('island', inplace=True)
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/titanic.csv')
```

```python
df1
```

```python
# This code removes columns from `df1` that contain any missing values (NaN).
# The `axis=1` argument specifies that columns should be dropped, not rows.
# `inplace=True` ensures the modification is applied directly to `df1`.
df1.dropna(axis=1)
```

```python
df1.fillna('ABC')
```

```python
# Grouping
g = df1.groupby('survived')
```

```python
g
```

```python
g.sum()
```

```python
# Previously, g.sum() produced concatenated strings because it summed all columns.
# Using `numeric_only=True` resolves this by summing only the numerical columns,
# avoiding the concatenation of non-numeric data.
g.sum(numeric_only=True)
```

```python
g.mean(numeric_only=True)
```

```python
g1 = df1.groupby(['survived','sex'])
```

```python
g1.sum(numeric_only=True)
```

```python
g2 = df1.groupby('pclass')
```

```python
g2.sum(numeric_only=True)
```

```python
g2.max(numeric_only=True)
```

```python
# This code first calculates the maximum value for each numeric column
# within each passenger class (grouped by 'pclass' in `g2`).
# The `numeric_only=True` argument ensures only numerical columns are considered.
# Finally, `.transpose()` is applied to swap the rows and columns,
# making it easier to compare the maximum values across different passenger classes (pclass 1, 2, 3).
g2.max(numeric_only=True).transpose()
```

```python
df3 = df1[['survived','pclass','sex']][0:5]
```

```python
df4 = df1[['survived','pclass','sex']][5:10]
```

```python
pd.concat([df3,df4])
```

```python
# This code concatenates `df3` and `df4` column-wise (axis=1).
# Because `df3` and `df4` have different index ranges (0-4 and 5-9 respectively),
# `pandas` aligns them by index. Where an index exists in one DataFrame but not the other,
# `NaN` (Not a Number) values are introduced to fill the missing entries.
pd.concat([df3,df4],axis=1)
```

```python
# Count the number of people who survived (1) and did not survive (0)
df1.groupby('survived').size()
```

```python
# This code groups the DataFrame `df1` first by 'pclass' (passenger class)
# and then by 'sex'. After grouping, `size()` is used to count the number
# of rows in each group, effectively showing the distribution of males and
# females within each passenger class.
df1.groupby(['pclass', 'sex']).size()
```

```python
# Group by 'pclass' and 'survived' to count the number of people who survived
# (1) or did not survive (0) within each passenger class.
df1.groupby(['pclass', 'survived']).size()
```
