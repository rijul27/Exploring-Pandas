```python
s = df['DepTime'][0:8]
```

```python
type(s)
```

```python
l = ['a','b','c','d','e','f','g','h']
```

```python
s1 = pd.Series(list(s),index=l)
print(s1)
# creates a new pandas Series. It takes the values from the existing Series s
# by converting them into a Python list using list(s). Then, it uses the list l
# (which contains ['a','b','c','d','e','f','g','h']) to assign custom labels as
# the index for this new Series. So, instead of the default numerical index,
# you get a Series with 'a' through 'h' as its index.
```

```python
pd.Series(s,index=l)

# When you use pd.Series(s, index=l), pandas tries to align the existing Series
# s with the new index l. Since s has a default numerical index (0, 1, 2, ...)
# and l has a different, custom string index ('a', 'b', 'c', ...), there are no
# common labels between the two indices. Because pandas attempts to match labels
# during this operation, and finding no matches, it fills the new Series with
# NaN (Not a Number) values. This might not always raise an explicit error, but
# it often results in a Series full of NaNs, which is typically not the desired
# outcome and can be perceived as an error or incorrect behavior.
```

```python
s3 = pd.Series([100,200,300,400,500], index = [2,4,5,6,1])
```

```python
s3
```

```python
s4 = pd.Series([4,5,6,7,8,9], index=[9,10,4,5,13,14])
```

```python
s4
```

```python
s5 =  pd.concat([s3, s4])
```

```python
s5
```

```python
# s5[4] is accessing elements from the pandas Series s5 using its index label.
# In this particular Series, the index label 4 appears more than once due to the
# concatenation of s3 and s4. When you index a Series with a label that has
# duplicates, pandas returns all entries corresponding to that label as a new
# Series. That's why the output shows both 200 (from s3) and 6 (from s4), both
# with the index 4.

s5[4]
```
